# B. Instantiate Data Context

```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

# C. Connect to A `Datasource`

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-car-sales-report'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'car-sales'
path_to_data = 'P2M3_gita_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

# D. Create an Expectation Suite

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-car-sales-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

## D.1 - Expectations

Lakukan minimal 7 Expectations yang didalamnya harus ada Expectation untuk:
- to be unique
- to be between min_value and max_value
- to be in set
- to be in type list
3 jenis Expectation yang berbeda yang tidak diajarkan pada lecture Week 2 Day 5 AM - Data Ethics & Data Validation

You can see, the Expectation `expect_column_values_to_not_be_null` on column `congestion_surcharge` return `False` with `63.27 %` of the rows are missing values.

```python
# Expectation 1 : Column `car_id` must be unique

validator.expect_column_values_to_be_unique('car_id')
```

```python
# Expectation 2 : Column `price` must between $1200 and $85800

validator.expect_column_values_to_be_between(
    column='price', min_value=1200, max_value=85800
)
```

```python
# Expectation 3 : Column `gender` must be in set ['Male', 'Female']

validator.expect_column_values_to_be_in_set('gender', ['Male', 'Female'])
```

```python
# Expectation 4 : Column `annual_income` in type list [int32, int64]

validator.expect_column_values_to_be_in_type_list('annual_income', ['int32', 'int64'])
```

```python
# Expectation 5 : Column 'date' has date format %Y-%m-%d

validator.expect_column_values_to_match_strftime_format('date', '%Y-%m-%d')
```

```python
# Expectation 6 : number of rows in the data set is 23906

validator.expect_table_row_count_to_equal(23906)
```

```python
# Expectation 7 : Column `body_style` must be exist
validator.expect_column_to_exist(column='body_style')
```

```python
# Expectation 8 : Column 'dealer_region' not contain non-ASCII characters

validator.expect_column_values_to_not_match_regex('dealer_region', r'[^\x00-\x7F]+')
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```
